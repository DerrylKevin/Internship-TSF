### Importing necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SampleSuperstore.csv")
```

### View the pairwise correlation

```python
cm = df.corr()
plt.figure(figsize=(11,8))
sns.heatmap(cm, annot=True)
```

### Plotting the categorical variable

```python
plt.style.use("seaborn")
```

#### Ship Mode

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Ship Mode', data=df)
plt.xlabel("Ship mode",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

#### Segment

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Segment', data=df)
plt.xlabel("Segment",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

#### Region

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Region', data=df)
plt.xlabel("Region",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

#### Category

```python
plt.figure(figsize=(7,4))
sns.countplot(x='Category', data=df)
plt.xlabel("Category",fontsize=15)
plt.ylabel("Count",fontsize=15)
plt.show()
```

#### Sub-Category

```python
plt.figure(figsize=(18,7))
sns.countplot(x='Sub-Category', data=df, order= df['Sub-Category'].value_counts().index)
plt.xlabel("Sub-Category", fontsize=15)
plt.ylabel("Count")
plt.show()
```

#### State

```python
plt.figure(figsize=(15,10))
states = sns.countplot(x='State', data=df, order= df['State'].value_counts().index)
states.set_xticklabels(states.get_xticklabels(), rotation=90)
plt.xlabel("State", fontsize=15)
plt.ylabel("Count", fontsize=15)
plt.show()
```

## Plotting pivot tables of categorical variables with respect to Sales and Profits

### 1. Ship Mode

```python
df.pivot_table(index="Ship Mode", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(5,4));
df.pivot_table(index="Ship Mode", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(5,4));
```

### 2. State

```python
df.pivot_table(index="State", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(16,5));
df.pivot_table(index="State", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(16,7))
```

### 3. Segment

```python
df.pivot_table(index="Segment", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(6,4));
df.pivot_table(index="Segment", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(6,4));
```

### 4. Region

```python
df.pivot_table(index="Region", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(6,4));
df.pivot_table(index="Region", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(6,4));
```

### 5. Category

```python
df.pivot_table(index="Category", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(6,4));
df.pivot_table(index="Category", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(6,4));
```

### 6. Sub-category

```python
df.pivot_table(index="Sub-Category", values="Sales", aggfunc='sum').sort_values(by="Sales", ascending=False).plot.bar(figsize=(10,5));
df.pivot_table(index="Sub-Category", values="Profit", aggfunc='sum').sort_values(by="Profit", ascending=False).plot.bar(figsize=(10,5));
```
